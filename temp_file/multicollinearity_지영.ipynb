### 다중공선성
- 독립변수 X는 종속변수 Y와의 상관관계만 있어야 하며, 독립변수들끼리는 상관 관계가 있으면 안됨
- 독립변수 간 상관관계가 보이는 것을 다중공선성 Multicollinearity
- 다중공선성이 있으면 부정확한 회귀결과 도출 가능성 존재

#### 다중공선성 검사

```python
# VIF Variance Inflation Factors 분산팽창요인
# 일반적으로 VIF가 10이 넘으면 다중공선성 있다고 판단
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif['VIF Factor'] = [variance_inflation_factor(train_x, i) for i in range(train_x.shape[1])]
vif['features'] = train_x.columns
vif = vif.sort_values('VIF Factor').reset_index(drop = True)
vif
```

```python
# 삭제할 행의 조건 : VIF가 10이 넘으면 다중공선성 있다고 판단
ten = (vif['VIF Factor'] > 10)

# 조건을 만족하지 않는 행들로 이루어진 새로운 데이터프레임 생성
res = vif[~ten]
type(res)
```

```python
# 추출된 col들만 train에 다시 반영
features = list(res['features'])
features, type(features)
```

```python
origin = ['ticker', 'firm', 'volume', 'open', 'high', 'low', 'close', 'adjustTrue']
y = ['target']
new_col = origin + features + y
new_col
```

```python
train = train[new_col]
```
